```python
# Importing the libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline

# Ignore harmless warnings 
import warnings 
warnings.filterwarnings("ignore")

# Set to display all the columns in dataset
pd.set_option("display.max_columns", None)

import pandasql as psql
```

```python
df = pd.read_csv(r"C:\Users\V V Madhava\contor solns\samhith project 1\Absenteeism_data.csv")
df.head()
```

```python
df.drop(['ID'], inplace=True, axis=1)
df.sample(5)
```

```python
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Date'] = df['Date'].dt.month

df.head()
```

```python
pd.unique(df['Education'])
df['Education'] = df['Education'].map({1:0, 2:1, 3:1, 4:1})
df.sample(10)
```

```python
conditions = [
    (df['Reason for Absence'] >= 0) & (df['Reason for Absence'] <= 14),
    (df['Reason for Absence'] >= 15) & (df['Reason for Absence'] <= 17),
    (df['Reason for Absence'] >= 18) & (df['Reason for Absence'] <= 21),
    (df['Reason for Absence'] >= 22) & (df['Reason for Absence'] <= 27)
]
choices = [1, 2, 3, 4]

# Use np.select to apply the conditions and replacements
df['Reason for Absence'] = np.select(conditions, choices, default=df['Reason for Absence'])
```

```python
df.sample(10)
```

```python
df.head()
```

```python
df.nunique()
```

```python
df['Absenteeism Time in Hours'] = np.where(df['Absenteeism Time in Hours'] > 3, 1, 0)
```

```python
df.sample(10)
```

```python
indepvar = []
for col in df.columns:
    if col!='Absenteeism Time in Hours':
        indepvar.append(col)
        
target = 'Absenteeism Time in Hours'
x = df[indepvar]
y = df[target]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler(feature_range=(0, 1))
x_train = mms.fit_transform(x_train)
x_train = pd.DataFrame(x_train)
x_test = mms.fit_transform(x_test)
x_test = pd.DataFrame(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```
